```python
from unstructured.partition.pdf import partition_pdf

output_path = "/content/"
file_path = '/content/1-s2.0-S1359645408006745-main.pdf'

# Reference: https://docs.unstructured.io/open-source/core-functionality/chunking
chunks = partition_pdf(
    filename=file_path,
    infer_table_structure=True,            # extract tables
    strategy="hi_res",                     # mandatory to infer tables

    extract_image_block_types=["Image","Table"],   # Add 'Table' to list to extract image of tables
    image_output_dir_path=output_path,   # if None, images and tables will saved in base64

    extract_image_block_to_payload=True,   # if true, will extract base64 for API usage

    chunking_strategy="by_title",          # or 'basic'
    max_characters=10000,                  # defaults to 500
    combine_text_under_n_chars=2000,       # defaults to 0
    new_after_n_chars=6000,

    # extract_images_in_pdf=True,          # deprecated
)
```

```python
# We get  types of elements from the partition_pdf function
set([str(type(el)) for el in chunks])
```

```python
# Each CompositeElement containes a bunch of related elements.
# This makes it easy to use these elements together in a RAG pipeline.

chunks[0].metadata.orig_elements
```

```python
# This is what an extracted image looks like.
# It contains the base64 representation only because we set the param extract_image_block_to_payload=True

elements = chunks[0].metadata.orig_elements
chunk_images = [el for el in elements if 'Image' in str(type(el))]
chunk_images[0].to_dict()
```

```python
# Get the images from the CompositeElement objects
def get_images_base64(chunks):
    images_b64 = []
    for chunk in chunks:
        if "CompositeElement" in str(type(chunk)):
            chunk_els = chunk.metadata.orig_elements
            for el in chunk_els:
                if "Image" in str(type(el)):
                    images_b64.append(el.metadata.image_base64)
    return images_b64

images = get_images_base64(chunks)
```

```python
import base64
from IPython.display import Image, display

def display_base64_image(base64_code):
    # Decode the base64 string to binary
    image_data = base64.b64decode(base64_code)
    # Display the image
    display(Image(data=image_data))

display_base64_image(images[8])
```

```python
# prompt: print all the image

for image in images:
  display_base64_image(image)
  print(20*'_')
```

```python
# prompt: store all the imgage from above line store in folder

import os
import base64
from PIL import Image
from io import BytesIO

def save_images_from_base64(images, output_dir="images"):
    """Saves images from a list of base64 strings to a specified directory.

    Args:
        images: A list of base64 encoded image strings.
        output_dir: The directory to save the images to.
    """

    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for i, image_data in enumerate(images):
        try:
            image = Image.open(BytesIO(base64.b64decode(image_data)))
            image_path = os.path.join(output_dir, f"image_{i}.png")
            image.save(image_path)
            print(f"Image {i} saved to {image_path}")
        except Exception as e:
            print(f"Error saving image {i}: {e}")

# Assuming 'images' list is already populated from the previous code
save_images_from_base64(images)
```
